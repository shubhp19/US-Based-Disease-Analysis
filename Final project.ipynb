```python
import requests
import pandas as pd

# URL of the JSON data
url = "https://data.cdc.gov/api/views/hk9y-quqm/rows.json?accessType=DOWNLOAD"

# Sending a request to the URL
response = requests.get(url)

# Parsing the JSON data
data = response.json()

# Creating a data frame from the data
df = pd.DataFrame(data['data'], columns=[col['name'] for col in data['meta']['view']['columns']])

# Displaying the DataFrame
df

```

# Data cleaning 

```python
# Convert 'Year' and 'Month' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')

# Rename 'COVID-19 Deaths' column to 'Deaths'
df.rename(columns={'COVID-19 Deaths': 'Deaths'}, inplace=True)

# Convert 'COVID-19 Deaths' and 'Number of Mentions' to numeric, coercing errors
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
df['Number of Mentions'] = pd.to_numeric(df['Number of Mentions'], errors='coerce')
```

```python
# Parse 'Data As Of', 'Start Date', and 'End Date' into datetime
df['Data As Of'] = pd.to_datetime(df['Data As Of'])
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])
```

```python
# Drop columns with all null values, ignoring errors if columns don't exist
df.drop(columns=['created_meta', 'updated_meta'], errors='ignore', inplace=True)

# Drop additional columns not needed for analysis, ignoring errors if columns don't exist
df.drop(columns=['id', 'meta', 'position'], errors='ignore', inplace=True)
```

```python
# Normalize text data for 'Condition', 'Condition Group', and 'State'
df['Condition'] = df['Condition'].str.lower()
df['Condition Group'] = df['Condition Group'].str.lower()
df['State'] = df['State'].str.title()
```

```python
# Fill missing 'Year' with the mode (most frequent value)
year_mode = df['Year'].mode()[0]
df['Year'].fillna(year_mode, inplace=True)
# Drop rows where 'COVID-19 Deaths' is NaN
df.dropna(subset=['Deaths'], inplace=True)

```

```python
# Check and remove duplicates
df.drop_duplicates(inplace=True)
```

### Monthly and Year Trend of COVID-19 Deaths

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Histogram for 'Month'
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.histplot(df['Month'], bins=12, color='skyblue', edgecolor='black')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.title('Histogram of COVID-19 Deaths by Month')
plt.xticks(range(1, 13))  # Assuming 'Month' is numbered 1-12
plt.show()


# Calculate the frequency (count) of each year
year_counts = df['Year'].value_counts().sort_index()

# Histogram for 'Year'
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.histplot(df['Year'], bins=len(year_counts), color='salmon', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram of COVID-19 Deaths by Year')

# Create a line plot that joins the high points of the histogram
# We use 'plot' method from pandas (which uses matplotlib under the hood)
year_counts.plot(kind='line', color='blue', marker='o', linewidth=2)

# For better x-axis labels
plt.xticks(sorted(df['Year'].unique()))

plt.show()
```

### COVID-19 Mortality Rate by Condition Group

```python
condition_group_stats = df.groupby('Condition Group').agg({'Deaths': 'sum', 'Number of Mentions': 'sum'}).reset_index()
condition_group_stats['Mortality Rate'] = (condition_group_stats['Deaths'] / condition_group_stats['Number of Mentions']) * 100

# Sort by mortality rate
condition_group_stats = condition_group_stats.sort_values(by='Mortality Rate', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Mortality Rate', y='Condition Group', data=condition_group_stats)
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Condition Group')
plt.title('COVID-19 Mortality Rate by Condition Group')
plt.show()
```

# Regression 

### Linear Regression Analysis:

```python
#pip install scikit-learn
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Filter out rows with missing values in COVID-19 Deaths and Number of Mentions
filtered_df = df.dropna(subset=['Deaths', 'Number of Mentions'])

# Convert columns to numeric, just to be sure 
filtered_df['Deaths'] = pd.to_numeric(filtered_df['Deaths'])
filtered_df['Number of Mentions'] = pd.to_numeric(filtered_df['Number of Mentions'])

# Perform linear regression on data
X = filtered_df[['Number of Mentions']]
y = filtered_df['Deaths']

#fitting the model
model = LinearRegression()
model.fit(X, y)

# Visualize the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5)
plt.plot(X, model.predict(X), color='red', linewidth=3)
plt.xlabel('Number of mention')
plt.ylabel('COVID-19 Deaths')
plt.title('Linear Regression: Deaths vs. Cases')
plt.show()

# Print regression coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```

### Regressional Analysis on Death and conditions

```python
import statsmodels.api as sm
import pandas as pd

# Assuming 'df' is your DataFrame

# Convert categorical variables to numerical representation
df['Condition Group'] = pd.Categorical(df['Condition Group']).codes
df['Age Group'] = pd.Categorical(df['Age Group']).codes

# Handling missing values in 'Deaths' column (if any)
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')

# Remove rows with missing values in any of the selected columns
df = df.dropna(subset=['Condition Group', 'Age Group', 'Deaths'])

# Multiple regression analysis
X = df[['Condition Group', 'Age Group']]  # Independent variables
X = sm.add_constant(X)  # Add a constant (intercept) term
y = df['Deaths']  # Dependent variable

# Fit the multiple regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

```python
from scipy.stats import f_oneway

# Hypothesis test to compare COVID-19 deaths among different condition groups
condition_groups = df['Condition'].unique()
data_by_condition = [df[df['Condition'] == condition]['Deaths'] for condition in condition_groups]

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(*data_by_condition)

# Print the results
print("ANOVA F-statistic:", f_statistic)
print("p-value:", p_value)
```

#### Clustering Analysis 

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Filter out entries where State is 'United States' and drop duplicates
filtered_unique_states_df = df[df['State'] != 'United States'].drop_duplicates(subset='State')

# Assuming that 'COVID-19 Deaths' is a column in your DataFrame,
# convert 'COVID-19 Deaths' to numeric, sorting, and getting the top 5 states
top_states_df = (
    filtered_unique_states_df.assign(
        Deaths=pd.to_numeric(filtered_unique_states_df['Deaths'], errors='coerce')
    )
    .sort_values(by='Deaths', ascending=False)
    .head(5)
)

# Perform k-means clustering on the top states
kmeans = KMeans(n_clusters=3)
# Assume 'Another Metric' is a column in your DataFrame; replace it with the actual column name
top_states_df['Cluster'] = kmeans.fit_predict(top_states_df[['Deaths', 'Number of Mentions']])

# Plotting the clusters for the top states
plt.figure(figsize=(10, 6))
plt.scatter(top_states_df['Deaths'], top_states_df['Number of Mentions'], c=top_states_df['Cluster'], cmap='viridis')

# Annotate each point with the state name
for idx, row in top_states_df.iterrows():
    plt.text(row['Deaths'], row['Number of Mentions'], row['State'], fontsize=9)

plt.title('Cluster Analysis of Top 5 Unique States Based on COVID-19 Deaths and Another Metric')
plt.xlabel('COVID-19 Deaths')
plt.ylabel('Number of Mentions')
plt.show()
```

#### Geoploting Covid 19 deathson United states map by states

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import requests

# Load the GeoJSON for US states from a local file
gdf = gpd.read_file('us_states.geojson')

# Fetch the COVID-19 data from the CDC API
response = requests.get('https://data.cdc.gov/api/views/hk9y-quqm/rows.json?accessType=DOWNLOAD')
covid_data = response.json()

# Extract relevant data into a DataFrame
columns = [col['name'] for col in covid_data['meta']['view']['columns']]
df = pd.DataFrame(covid_data['data'], columns=columns)

# Convert 'COVID-19 Deaths' to numeric and handle missing data
df['COVID-19 Deaths'] = pd.to_numeric(df['COVID-19 Deaths'], errors='coerce')
df['COVID-19 Deaths'].fillna(0, inplace=True)

# Aggregate data by state
state_data = df.groupby('State')['COVID-19 Deaths'].sum().reset_index()

# Use the 'NAME' column from gdf for merging
merged = gdf.set_index('NAME').join(state_data.set_index('State'))

# Plot the choropleth map with an enhanced colormap
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged.plot(column='COVID-19 Deaths', ax=ax, legend=True, cmap='viridis',
            legend_kwds={'label': "COVID-19 Deaths by State",
                         'orientation': "horizontal"})

# Zoom in on the continental US
ax.set_xlim(-130, -65)
ax.set_ylim(24, 50)

plt.show()
```
